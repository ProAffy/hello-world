c. The steps of hierarchical clustering is as follows:
1. Data Input - It is the source from where data is collected.
2. Distance Metric Selection - Which distance metric to choose from, that would be appropriate is important. For ex- Eucledian Distance or Manhattan Distance
3. Distance matrix computation - Once the distance method is selected, we need to compute the distance matrix, that consists of pairwise distance between all data points.
4. Creating linkage matrix - Linkage matrix specifies that how the clusters will be merged. Some of linkage matrix include, single linkage and complete linkage.
5. Hierarchical Clustering - We can merge clusters based on the distance. The two closest data points can be kept in one cluster and so on. Continue the process until all the data points are in the single cluster.
6. Visualization of clusters - We can use dendrogram to visualize the clusters to depict hierarchy.

a. The Python code for implementing K-Means clustering algorithm to the dataset is below. Specific comments have been provided to make the code readable.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Loading the dataset
data = pd.read_csv('faithful.csv')

# Initializing the KMeans algorithm
kmeans = KMeans(n_clusters=2)

# Fitting the model to the dataset
kmeans.fit(data)

# Getting the centroids of the clusters
centroids = kmeans.cluster_centers_

# Getting the labels of the data points for display
labels = kmeans.labels_

# Creating a scatter plot of the data points colored according to their cluster labels
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels)

# Adding the cluster centers to the plot
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='#050505')

# Add axis labels and a title
plt.xlabel('Eruptions')
plt.ylabel('Waiting')
plt.title('KMeans Clustering')

# Show the plot
plt.show()
```

c. The plot of the k-means objective function versus iterations of the algorithm is depicted below. 

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('faithful.csv')

# Initialize KMeans algorithm with the desired number of clusters
kmeans = KMeans(n_clusters=2)

# Fit the model to the data and get the objective function for each iteration
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=3, max_iter=i, n_init=1)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

# Plot the objective function versus the number of iterations
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Iterations')
plt.ylabel('Objective Function')
plt.title('KMeans Clustering')

# Show the plot
plt.show()
```
